```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('company_sales_data.csv')
```

```python
plt.plot(data['month_number'], data['total_profit'], marker="o")
plt.title('Total Profit by month')
plt.xlabel('month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
fig, axs = plt.subplots(2, figsize=(10, 8))
axs[0].plot(data['month_number'], data['bathingsoap'], marker='o')
axs[0].set_title('Bathing Soap Sales by Month')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Sales')
axs[0].grid(True)

axs[1].plot(data['month_number'], data['facewash'], marker='o')
axs[1].set_title('Face Wash Sales by Month')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Sales')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
